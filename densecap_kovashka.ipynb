```python
import json
import os
import tqdm
```

```python
captions_densecap_test = json.load(open('results.json', 'r'))
captions_test = {}
```

```python
for image_path in tqdm.tqdm(captions_densecap_test.keys()):
    caption_set = set(captions_densecap_test[image_path])

    caption_string = ""

    for cap in caption_set:
      caption_string += cap + ". "

    captions_test[image_path] = caption_string
```

```python
print(len(captions_test))
```

```python
captions_test_output_file = open(os.path.join("test-set/test", "Captions-densecap_test.json"), 'w')
json.dump(captions_test, captions_test_output_file, indent=1)
```
